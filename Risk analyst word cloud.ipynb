```python
legitscript="Role Purpose: The purpose of this role is to accurately analyze assigned merchants, assess associated risk, and convey merchant analysis in an actionable note to the client, in accordance with program standards and deadline requirements. A risk analyst must develop a foundational knowledge of, as well as ability to make decisions in, all content areas monitored by the program. At this level, the employee is building foundational skills in order to become a full-fledged contributor to program work. There is an expected training period for all analysts (currently 60-90 days). Over the course of the first year at LegitScript, an analyst should be fully contributing to the program(s). Key Area #1: Culture and values alignment Engage in behaviors aligned with LegitScript’s core values: Be Awesome, Be Open, Be Thoughtful, Be Invested Exhibit a growth mindset by being open to feedback, embracing change as an opportunity Key Area #2: Accurately analyze and report merchants Analyze and assess merchant risk levels by using internet investigative techniques and leveraging contextual data Make quick, accurate analytical decisions based on information that is, at times, ambiguous Provide analysis details to clients under strict deadlines Adhere to program’s reporting standards Key Area #3: Gain expertise in high-risk merchant business models Attain and maintain a thorough understanding of regulations and policies for various high-risk areas in the payment industry Complete training and attend team meetings Stay informed of market trends by reading relevant resources shared by team members Required skills Alignment with company values and culture Growth mindset (receptive to feedback, learning) Detail-oriented Independent/resourceful Efficient time management Self-motivated Curious Ability to work under pressure Tech savvy Adaptable Good judgment/common sense Clear communicator (both in writing and verbally) Bachelors degree or equivalent experience Investigative research/analysis skills Experience with large data sets (Excel not necessary) Experience assessing risk Preferred skills 1-2+ years work, especially in relevant field Management experience (project or people) Bilingual Relevant specialty skill(s) [e.g., writing/editing, client relations]"
```

```python
stopwords = set(STOPWORDS)
```

```python
legitscript_wc = WordCloud(
    background_color='lightyellow',
    max_words=2000,
    stopwords=stopwords)

legitscript_wc.generate(legitscript)
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 


plt.imshow(legitscript_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
